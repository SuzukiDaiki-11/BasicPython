```python
# 使うライブラリ
# !pip install japanize-matplotlib

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import japanize_matplotlib
japanize_matplotlib.japanize()
```

## 1問目
使うcsvはhist_data.csvです。下にコードを書いてください。

![問1](./imgs/matplotlib/plot_1.png)

```python
data = pd.read_csv('data/hist_data.csv')

data_a = data['aデータ']
data_b = data['bデータ']

plt.hist(data_a, alpha = 0.5, bins = 30, label = 'aデータ', color = 'orange')
plt.hist(data_b, alpha = 0.5, bins = 30, label = 'bデータ', color = 'green')

plt.xlabel('温度(℃)')
plt.xticks([10, 20, 30, 40, 50])
plt.ylabel('度数')
plt.yticks([5, 10, 15, 20, 25])
plt.title("aデータとbデータの温度に関する度数分布")

plt.legend()
plt.grid(False)

plt.show()
```

## 2問目
使うcsvはscatter_data.csvです。下にコードを書いてください。

![問2](./imgs/matplotlib/plot_2.png)

```python
data = pd.read_csv('data/scatter_data.csv')

class0 = data[data['class'] == 0]
class1 = data[data['class'] == 1]
class2 = data[data['class'] == 2]

plt.scatter(class0['length (cm)'], class0['width (cm)'], label = 'class0', color = "r")
plt.scatter(class1['length (cm)'], class1['width (cm)'], label = 'class1', color = "b")
plt.scatter(class2['length (cm)'], class2['width (cm)'], label = 'class2', color = "g")

plt.xlabel('lenght (cm)')
plt.xticks([4.5, 5.0, 5.5, 6.0, 6.5, 7.0, 7.5, 8.0])
plt.ylabel('width (cm)')
plt.yticks([2.0, 2.5, 3.0, 3.5, 4.0, 4.5])
plt.title('クラス0~2のlengthとwidthに関する散布図')

plt.legend()
plt.legend()
plt.grid(True)

plt.show()
```

## 3問目
使うcsvはline_data.csvです。下に2つの画像を表示するコードを書いてください。

![問3_1](./imgs/matplotlib/plot_3_1.png)

![問3_2](./imgs/matplotlib/plot_3_2.png)

```python
data = pd.read_csv('data/line_data.csv')

plt.figure(figsize=(10, 5))
data.set_index('期間', inplace=True)
data_a = data['件数']
plt.plot(data_a, marker='o')

plt.xlabel('期間')
plt.ylabel('件数')
plt.title('期間ごとの件数の推移')
plt.legend()
plt.grid(True)

plt.show()
```

```python
data = pd.read_csv('data/line_data.csv')

period = data['期間']
data_b = data['費用']

plt.bar(period, data_b)

plt.xlabel('期間')
plt.xticks(rotation = 30)
plt.ylabel('費用 (円)')
plt.title('期間ごとの費用の推移')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```
